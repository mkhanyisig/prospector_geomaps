```python
# load libraries
import pandas as pd
import csv
import json
from IPython.display import display
import time

#Goal: add mineral tags for each location
```

## Load JSON file and create object

```python
with open('usa_test_2000.json') as f:
  json_obj = json.load(f)
#print(type(json_obj))
print("****\nTesting \n****\n")
print("Total Locations: ",len(json_obj['features']))
print("\nGeometry:\n",json_obj['features'][0]['geometry'])
```

### Create and add tags

```python
minerals=set()

i=0
for location in json_obj['features']: # first 10, replace for general later
    #print(location['properties']['commod1']," * ",location['properties']['commod2'],' * ',location['properties']['commod3'])
    min1=location['properties']['commod1']
    min2=location['properties']['commod2']
    min3=location['properties']['commod3']
    
    mineral_tags=set()
    
    print("\n***\n")
    # first tag
    if min1!=None:
        print(min1)    
        min1=min1.split(",")
        temp=[]
        for el in min1:
            el=el.replace(" ","")
            el=el.lower()
            temp.append(el) 
            mineral_tags.add(el)
        print(temp)
    # second tag
    if min2!=None:
        print(min2)    
        min2=min2.split(",")
        temp=[]
        for el in min2:
            el=el.replace(" ","")
            el=el.lower()
            temp.append(el) 
            mineral_tags.add(el)
        print(temp)
    # second tag
    if min3!=None:
        print(min3)    
        min3=min3.split(",")
        temp=[]
        for el in min3:
            el=el.replace(" ","")
            el=el.lower()
            temp.append(el) 
            mineral_tags.add(el)
        print(temp)
        
    print("Minerals: ",list(mineral_tags))
    mineral_tags=list(mineral_tags)
    json_obj['features'][i]['properties']['mineral_tags']=mineral_tags
    #print(json_obj['features'][i]['properties'])
    i+=1
    
```

## Export JSON Object

```python
print("Sample Feature Object:\n\n",json_obj['features'][0])
```
